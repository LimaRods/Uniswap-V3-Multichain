```python
df_c = pd.read_csv('Data Sources/v3-cohort-analysis-eth.csv')
#Retention
cohort = df_c.pivot(index = 'cohort_month', columns = 'month_number', values = 'num_users').fillna(0)
#columns = cohort.columns
#for col in columns:
    #cohort[col] = cohort[col].apply(lambda x: feat.human_format(x))
    
# New users who returned
cohort_pct = round(df_c.pivot(index = 'cohort_month', columns = 'month_number', values = 'retention'),2).fillna(0)
```

```python
# New users who returned

#visualization
fig, ax = plt.subplots()
sea.heatmap(cohort,annot = True, fmt=  "0.0f" ,cmap = "Blues")


#Setting
plt.title("Cohort Matrix - New Users Who Returned MoM between 2021 and 2022 for Uniswap on Ethereum", fontweight = 'bold', fontsize = 12)
plt.gcf().set_size_inches(20,15)
plt.gcf().set_dpi(100)
#plt.subplots_adjust(top = .99)
plt.xticks(size = 15)
plt.yticks(size = 12)
ax.xaxis.set_ticks_position('top')

plt.show()
```

```python
# New users who returned

#visualization
fig, ax = plt.subplots()
sea.heatmap(cohort_pct.iloc[:,1:],annot = True,cmap = "Blues")


#Setting
plt.title("Cohort Matrix - Retention Rate MoM between 2021 and 2022 for Uniswap on Ethereum", fontweight = 'bold', fontsize = 12)
plt.gcf().set_size_inches(20,15)
plt.gcf().set_dpi(100)
#plt.subplots_adjust(top = .99)
plt.xticks(size = 15)
plt.yticks(size = 12)
ax.xaxis.set_ticks_position('top')

plt.show()
```
